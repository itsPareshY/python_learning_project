# Python Installation Guide
This notebook contains steps to install Python, pip, and required packages for the Python Learning Project.

## Step 1: Install Python
1. Download the Python installer from the official website: [python.org](https://www.python.org/downloads/)
2. Run the installer and follow the instructions. Make sure to check the box that says 'Add Python to PATH'.
3. Verify the installation by running the following command in your command prompt:
   
    ```bash
   python --version
   ```
   You should see the installed Python version.


## Step 2: Install pip
1. Pip is included by default with Python 3.4 and later. If it is not installed, download the script from [get-pip.py](https://bootstrap.pypa.io/get-pip.py).
2. Run the script using the following command:
   
    ```bash
   python get-pip.py
   ```
3. Verify the installation by running:
   
    ```bash
   pip --version
   ```
   You should see the installed pip version.


### Step 2.1: Upgrade pip
1. You can upgrade pip to the latest version using the following command:
   ```bash
