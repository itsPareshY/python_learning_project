# Working with Collections

## 1. Lists

Lists are ordered, mutable collections that can hold a variety of data types.

You can create a list using **square brackets []**.

Example:

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0])  # Output: app
```

````python
fruits = ["apple" , "banana", "mango"]

print(fruits[0]) # print item at specific index
print(fruits)

fruits.append("papaya") # add items to end of list
print(fruits[3])

print(f"fruits is of data type {type(fruits)}")

fruits.remove("banana") # remove items from list 
print(fruits) # banana removed from list 

````

## 2. Tuples

Tuples are similar to lists but are immutable (cannot be changed).

You can create a tuple using parentheses **()**.

Example:

```python
coordinates = (10.0, 20.0)
print(coordinates[1])  # Output: 20.0
```

````python

fruit_price = ("cherry", 20.0)
print(fruit_price[0])
print(fruit_price[1])
````

## 3. Dictionary

Dictionaries are unordered collections of **key-value** pairs.

You can create a dictionary using **curly braces {}**.

Example:
```python
student = {"name": "Alice", "age": 25, "grades": [85, 90, 78]}
print(student["name"])  # Output: Alice
```

````python
student = {"name": "Alice", "age": 25, "grades": [85,80,78]} 
print(student["name"])
````

````python
ice_cream_flavors = {
    "Mint Chocolate Chip": "Refreshing mint ice cream studded with decadent chocolate chips.",
    "Cookie Dough": "Vanilla ice cream loaded with chunks of chocolate chip cookie dough.",
    "Salted Caramel": "Sweet and salty with a smooth caramel swirl and a hint of sea salt."
}   
````

````python
print(ice_cream_flavors)
````

##### Get Value for a known key

````python
cookie_dough_description = ice_cream_flavors["Cookie Dough"] # get specific key value
print(cookie_dough_description)
````

##### Print all keys in dictionary

````python
print(ice_cream_flavors.keys()) # Print keys
````

##### Print all values in dictionary

````python
print(ice_cream_flavors.values()) # Print values
````

##### Add/Update in dictionary

````python
ice_cream_flavors["Rocky Road"] = "Chocolate ice cream mixed with other ingredients." # Add / Update item in dictionary
````

##### Print value of specific key

````python
print(ice_cream_flavors)
````

##### Remove known key in dictionary

````python
del ice_cream_flavors["Rocky Road"]  # Removes the key-value pair with key "Rocky Road"
````

````python
print(ice_cream_flavors)
````

##### pop() function to remove specific key from  dictionary and return its value

````python
my_dict = {"a": 1, "b": 2, "c": 3}
value = my_dict.pop("b")  # Removes 'b' and returns its value
print(my_dict)  # Output: {'a': 1, 'c': 3}
print(value)    # Output: 2
````

##### popitem() - remove random item and return its value

````python
my_dict = {"a": 1, "b": 2, "c": 3}
item = my_dict.popitem()  # Removes a random key-value pair
print(my_dict)  # Output: {'a': 1, 'b': 2} (or another random pair is removed)
print(item)     # Output: ('c', 3) (or another random pair)
````

##### Clear dictionary

````python
my_dict = {"a": 1, "b": 2, "c": 3}
my_dict.clear()  # Clears all items
print(my_dict)   # Output: {}
````

##### Remove specific keys example

````python
my_dict = {"a": 1, "b": 2, "c": 3, "d": 4}

# List of keys to remove
keys_to_remove = ["b", "d"]

# Removing keys using a loop
for key in keys_to_remove:
    my_dict.pop(key, None)  # None prevents KeyError if the key is not present

print(my_dict)  # Output: {'a': 1, 'c': 3}
````

## 4. Sets

Sets are unordered collections of unique elements.

You can create a set using **curly braces {}** or the **set()** function.

Example:
```python
unique_numbers = {1, 2, 3, 3, 4}  # Duplicates are removed
print(unique_numbers)  # Output: {1, 2, 3, 4}
```

````python
unique_numbers = {6,7,9,2,1,2,3,3,3,3,4,5,5}
print(unique_numbers)
# print(unique_numbers[3]) -> cannot access elements in set using subscripting, instead convert to list , or iterate 

#convert set to list to access elements by index
my_list = list(unique_numbers)
print(f"\nprinting 3rd element in list {my_list[2]}\n")

#Iterate set elements
for element in unique_numbers:
    print(element)
````

# Combination of Collections

## 1. List of Tuples

A list where each element is a tuple.


````python
list_of_tuples = [(1, 'apple'), (2, 'banana'), (3, 'cherry')]
print(list_of_tuples)
````

## 2. Tuples of List

A tuple where each element is a list.

````python
tuples_of_list = (["apple","banana","mango"],["potato","onion","tomato"])
print(tuples_of_list)
````

## 3. List of Dictionaries

A list where each element is a dictionary.

````python
list_of_students = [
    {"name": "Paresh", "class": 10, "grades":(8.2,7.5,6.1)},
    {"name": "Raghav", "class": 10, "grades":(9.2,8.5,7.1)}
]
print(list_of_students)
````

## 4. Dictionary of Lists

````python
dict_of_lists = {
    "fruits": ["apple","banana","cherry"],
    "vegetables": ["potato", "onion","cabbage"],
    "nuts": ["almond","cashew","walnut"]
}
print(dict_of_lists)
````

## 5. Set of Dictionaries

A set where each element is a dictionary. Note that dictionaries must be immutable to be added to a set, so you can use "frozenset" for this purpose:

````python
set_of_dicts = {
    frozenset({'name': 'Alice', 'age': 25}.items()),
    frozenset({'name': 'Bob', 'age': 30}.items())
}
print(set_of_dicts)
````

## 6. Dictionary of Sets
A dictionary where each key maps to a set can be created as follows:

````python
dict_of_sets = {
    'fruits': {'apple', 'banana', 'cherry'},
    'vegetables': {'carrot', 'lettuce', 'spinach'}
}
print(dict_of_sets)
````
