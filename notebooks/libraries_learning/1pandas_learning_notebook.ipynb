# Learning Pandas

**Pandas is a powerful library for data manipulation and analysis. It's widely used for handling structured data and is particularly useful in data science.**

This notebook will cover the following topics in Pandas:

1. Data Structures:
   - Series : 1D array-like object
   - DataFrame : Two-dimensional labeled data structure with columns of potentially different types.

2. Data Manipulation:
   - Reading and writing data (CSV, Excel, SQL)
   - Data cleaning and preprocessing (handling missing values, removing duplicates, etc.)
   - Merging and joining datasets

3. Data Analysis:
   - Descriptive statistics
   - Grouping data

4. Data Visualization:
   - Basic plotting capabilities
