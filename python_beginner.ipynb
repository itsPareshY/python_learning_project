## File Handling

#### with open('file.txt', 'r') as file:
This line opens the file named file.txt in read mode ('r').
The with statement is used to ensure that the file is properly closed after its suite finishes, even if an error occurs. This is a context manager that handles file opening and closing automatically.
The variable file is a file object that allows you to interact with the file.
#### content = file.read()
This line reads the entire content of the file and stores it in the variable content.
The read() method reads the file's contents as a single string. If you want to read the file line by line, you could use file.readlines() instead.

````python
# Create or open the file 'file.txt' and write some content to it
with open('file.txt', 'w') as file:
    file.write("Hello, world!\nThis is a test file.\n")
    file.write("Hello, Mumma!\n")
    file.write("This is a lines test.\n")

# Now read the content of the file
with open('file.txt', 'r') as file:
    content = file.read()  # Read the entire content
    print(content)  # Print the entire content

    # Move the file pointer back to the start of the file
    file.seek(0)  # Reset the pointer to the beginning of the file
    lines = file.readlines()  # Now read the lines

# Printing each line
for line in lines:
    print(line.strip())  # Using strip() to remove the newline character
````

```python

```    
The issue with print(line.strip()) not printing anything is due to how you are reading the file. When you call file.read() before file.readlines(), the file pointer moves to the end of the file, leaving nothing to read when you subsequently call readlines().

1. File Reading Methods:
file.read() reads the entire content of the file and moves the file pointer to the end.
file.readlines() reads the lines from the current position of the file pointer.

## F Strings

In Python, f"..." and f'''...''' are used for f-strings, which are a way to format strings using expressions inside curly braces {}. This feature was introduced in Python 3.6 and provides a more readable and concise way to include variables and expressions in strings.

Usage of f-strings

1. Basic f-string:

You can include variables directly in the string:


````python
name = "Alice"
greeting = f"Hello, {name}!"
print(greeting)  # Output: Hello, Alice!
````

2. Expressions in f-strings:
    You can also include more complex expressions:

````python
x = 5
y = 10
result = f"The sum of {x} and {y} is {x + y}."
print(result)  # Output: The sum of 5 and 10 is 15.
````

3. Multiline f-strings:
    You can use triple quotes for multiline f-strings:


````python
name = "Alice"
age = 30
info = f"""Name: {name}
Age: {age}"""
print(info)
````

Use f"..." for single-line f-strings and f'''...''' for multiline f-strings.
